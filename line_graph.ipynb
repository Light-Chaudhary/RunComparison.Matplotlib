```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020]

kohli_runs = [850, 900, 1000, 1100, 950, 1050]
rohit_runs = [700, 850, 900, 950, 1000, 1100]
dhawan_runs = [650, 700, 750, 800, 850, 900]

plt.style.use("ggplot")

plt.plot(years, kohli_runs, color = "red", linestyle = "-", marker = "*", label = "Kohli Run")
plt.plot(years, rohit_runs, color = "blue", linestyle = "-", marker = "*", label = "Rohit Run")
plt.plot(years, dhawan_runs, color = "green", linestyle = "-", marker = "*", label = "Dhawan Run")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.grid(True)
plt.legend()
plt.xticks(years)
plt.show()
```
